```python
import pandas as pd
```

```python
data = pd.read_csv('../data_2/mhealth_raw_data.csv')
data
```

```python
columns = data.columns.to_list()
len(columns)
```

```python
# The activity label 0 corresponds to recordings in between the exercises. Thus, discard those instances
data = data[data['Activity'] != 0]
data
```

```python
# train test split
from sklearn.model_selection import train_test_split
train_percentage = 0.7
test_percentage = 0.3
```

```python
grouped = data.groupby(['subject', 'Activity'])

train_data = []
test_data = []
```

```python
for group, group_data in grouped:
    train, test = train_test_split(group_data, test_size=test_percentage, random_state=42)
    train_data.append(train)
    test_data.append(test)
```

```python
train_data = pd.concat(train_data)
test_data = pd.concat(test_data)
```

```python
train_data.shape, test_data.shape
```

```python
# save train and test data to csv
train_data.to_csv('../data_prep/train_data.csv', index=False)
test_data.to_csv('../data_prep/test_data.csv', index=False)
```
